```python
#We will have to load our file as a NLTK Text object inorder to make use of some helpful functions.
import nltk.corpus
from nltk.text import Text

vcs_speech=Text(nltk.corpus.gutenberg.words('C:\\Users\\Samuel Theophilus\\PycharmProjects\\nlp-101.txt'))

vcs_speech.concordance('institutions')
```

```python
#find list of words similar in context to the input
vcs_speech.similar('wesley')
```

```python
#find the contexts that are shared by these words in the text file
vcs_speech.common_contexts('achievements', 'recognition')
```

```python
from nltk.draw.dispersion import dispersion_plot

#plot a graph to view the repetition of some words in the text file
dispersion_plot(vcs_speech,['Wesley','University','Nigeria','success','The','the'])
```

```python
#pull a set of all words and punctions used in the document.
sorted(set(vcs_speech))
```

```python
#count the number f occurence of a word
print(vcs_speech.count('the'))

#count total amount of words
print(len(vcs_speech))

#Slice Text
print(vcs_speech[1430:])
```

```python
#Find the Frequency Distribution
fdist= FreqDist(vcs_speech)

vocabulary_fdist=list(fdist.keys())
#get a list(not set) of words in Text
print(vocabulary_fdist[:5])
#Visualize the First 20 most reocurring words
fdist.plot(20)
```

```python
#lets try and create a list of all words long than 8 characters in the Given Text
long_words = [word for word in vcs_speech if len(word) > 8]
print(long_words)
```
